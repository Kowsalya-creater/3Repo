```python
#importing from libraries
import pandas as pd
import pickle
```

```python
# converting csv file to dataFrame
df = pd.read_csv('heart_data.csv')
df.head()
```

```python
# Deleting rows with null value
df.dropna(inplace = True)
df.head()
```

```python
# Seperates and stores all the column EXCEPT 'target - column' under the variable input
input = df.drop('target', axis = 1)
input.head()
```

```python
#Seperates the column 'target' from the data and stores it under te variable output
output = df['target']
output.head()
output.tail()
```

```python
# importing this for spliting data into 4 - input training, input testing, output training, output testing 
from sklearn.model_selection import train_test_split
```

```python
# spliting 20% of the data for testing and the rest 80% goes to training
# random_state number 42(is a popupar number, can give any number)
ip_train, ip_test, op_train, op_test = train_test_split(input, output, test_size=0.2, random_state= 42)
print(ip_train.head())
print(ip_test.head())
```

```python
# importing the Model -Decision tree and storing it in a short variable(d_tree)
#created the decision tree model
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier(random_state = 42)
d_tree.fit(ip_train, op_train)
```

```python
# tried predicting b4 saving
d_tree_prediction = d_tree.predict(ip_test)
print(d_tree_prediction)
```

```python
#saving the model as pickle file
with open('DecisionModel.pkl', mode = 'wb') as d:
    pickle.dump(d_tree,d)
```

```python
# after saving the model we have to evaluate the model, before which we have to READ THE FILE
with open('DecisionModel.pkl', mode = 'rb') as r:
    D_tree_model = pickle.load(r)
```

```python
#Making the model predict the output with the training given with our previous data
d_tree_prediction = d_tree.predict(ip_test)
print(d_tree_prediction)
```

```python
# importing this to cross check the result between actual output given in the data and the predicted data by the model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#checking the accuracy of the model's output with the data's output
d_model_accuracy = accuracy_score(op_test, d_tree_prediction)
print(f"Accuracy: {d_model_accuracy * 100:.2f}%")
```

```python
classi_rep_accuracy = classification_report(op_test, d_tree_prediction)
print(classi_rep_accuracy)
```
