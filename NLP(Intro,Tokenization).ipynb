Tokenization
1.Using NLTK (Natural Language Toolkit):


```python
import nltk
nltk.download('punkt')  # Download the tokenizer data
from nltk.tokenize import word_tokenize

text = "Natural Language Processing is fascinating!"
tokens = word_tokenize(text)

print(tokens)
```

# 2. Using spaCy:

```python
import spacy

# Load the spaCy model for English
nlp = spacy.load("en_core_web_sm")

text = "Natural Language Processing is fascinating!"
doc = nlp(text)

tokens = [token.text for token in doc]

print(tokens)
```

# Subword Tokenization
1. Byte-Pair Encoding (BPE)


```python
from transformers import GPT2Tokenizer

# Load GPT-2 tokenizer (uses Byte-Pair Encoding)
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

text = "I enjoy learning about Natural Language Processing"

# Tokenize text using BPE
tokens = tokenizer.tokenize(text)

print(tokens)
```

2. **WordPiece**

```python
from transformers import BertTokenizer

# Load HuggingFace BERT tokenizer (uses WordPiece)
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

text = "I enjoy learning about Natural Language Processing"

# Tokenize text using WordPiece
tokens = tokenizer.tokenize(text)

print(tokens)
```

Stop words
1. NLTK Stop Words Example:

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Get the English stop words list
stop_words = set(stopwords.words('english'))

# Sample text
text = "This is a sample sentence, showing off stop words filtration."

# Tokenize the text
word_tokens = nltk.word_tokenize(text)

# Remove stop words
filtered_sentence = [word for word in word_tokens if word.lower() not in stop_words]

print("Original:", word_tokens)
print("Filtered:", filtered_sentence)
```
