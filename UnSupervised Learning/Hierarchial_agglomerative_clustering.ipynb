```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# using nested list coz, each datapoint is an individual cluster in the begining
data_points = [[18], [22], [25], [27], [42], [43]]
```

```python
z = linkage(data_points, method = 'ward')
```

```python
#visualize
plt.figure(figsize = (8,5))
```

```python
#create dendrograms to visualize
dendrogram( z, labels =[18, 22, 25, 27, 42, 43])
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel("Data Points")
plt.ylabel('Distance')
plt.show()
```
