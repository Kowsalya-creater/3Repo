```python
#import necessary library
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
#load the iris dataset
iris = load_iris()
data = iris.data
```

```python
# apply K-means clustering
kmeans = KMeans (n_clusters= 3, random_state =42)
kmeans.fit(data) # directly fitting, we are not training the dataset here, like how we do in supervised learning
clusters = kmeans.labels_  # 3 clusters - 1,2 3
```

```python
# convert to dataframe for easier plotting
iris_df = pd.DataFrame(data, columns = iris.feature_names)
print(iris_df.head())
iris_df['Cluster'] = clusters
print(iris_df)
```

```python
#plotting
plt.figure(figsize = (10,6))
```

```python
#use sepal length and sepal width for visualizaton
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c = iris_df['Cluster'], cmap = 'viridis', s=50)

# Mark cluster centers
centers = kmeans.cluster_centers_ #randomly initiatlises the cluster center
plt.scatter(centers[:,0], centers[:,1], c = 'red', s = 200, alpha = 0.75, marker = 'X')

# labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-means Clustering on iris Dataset')
plt.show()
```
