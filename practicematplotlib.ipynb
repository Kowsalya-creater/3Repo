```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv("heart_disease.csv")
y=pd.DataFrame(x)
```

```python
print(y.head())
```

```python
y.dropna(subset="chol",inplace=True)
print(y)
```

```python
y.dropna(subset="age",inplace=True)
print(y)
```

```python
x_axis=y["age"]
y_axis=y["chol"]

plt.plot(x_axis,y_axis)
plt.show()

```

```python
plt.scatter(x_axis,y_axis,color='red')
plt.xlabel('age')
plt.ylabel('chol')
plt.show()
```

```python
a=pd.read_json("sample.json")
df=pd.DataFrame(a)
print(df.head())
```

```python
df["age"].fillna(df["age"].mean())
```

```python
df.dropna(subset="salary",inplace=True)
```

```python
x_axis=df["age"]
y_axis=df["salary"]
plt.scatter(x_axis,y_axis)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("age Vs salary")
#plt.plot(x_axis,y_axis)
plt.show()
```
